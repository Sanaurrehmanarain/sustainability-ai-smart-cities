```python
import os
import yaml
from pathlib import Path

# ==========================================
# Configuration: Project Structure Definition
# ==========================================
PROJECT_NAME = "Multi-Modal-Sustainability-AI"

# Define the directory structure
structure = {
    "data": ["raw", "processed", "external"],
    "notebooks": [],
    "src": ["data", "features", "models", "visualization"],
    "models": ["checkpoints", "final"],
    "reports": ["figures", "sections"], # For the MS Word report parts
    "tests": [],
    "config": [],
    "docker": []
}

# Define initial files and their content
files = {
    "README.md": f"# {PROJECT_NAME}\n\n## Multi-Modal 'Sustainability AI' for Smart Cities\n\nExecutive summary and project description will go here.",
    "TASKS.md": "# Project Tasks & Roadmap\n\n- [x] Project Setup\n- [ ] Data Acquisition (Sentinel-2 & Census)\n- [ ] Preprocessing Pipeline\n- [ ] Modeling (CNN/ViT + XGBoost)\n- [ ] Evaluation & Explainability\n- [ ] Reporting\n",
    "requirements.txt": "numpy\npandas\nmatplotlib\nseaborn\nscikit-learn\ntorch\ntorchvision\nxgboost\nlightgbm\nshap\nearthengine-api\ngeemap\nrasterio\npytest\npyyaml\n",
    ".gitignore": "__pycache__/\n*.pyc\n.ipynb_checkpoints/\n.env\ndata/\nmodels/\n!data/.gitkeep\n!models/.gitkeep\n",
    "config/config.yaml": """
project:
  name: "Sustainability-AI-Smart-Cities"
  seed: 42

data:
  raw_path: "data/raw"
  processed_path: "data/processed"
  satellite_source: "Sentinel-2"
  
model:
  cnn_backbone: "resnet50" # or vit_b_16
  tabular_model: "xgboost"
""",
    "Dockerfile": """
# Base Image
FROM python:3.9-slim

# Working Directory
WORKDIR /app

# Copy Requirements
COPY requirements.txt .

# Install Dependencies
RUN pip install --no-cache-dir -r requirements.txt

# Copy Source Code
COPY src/ ./src/

# Command
CMD ["python", "src/main.py"]
"""
}

# ==========================================
# Execution: Create Directories and Files
# ==========================================
def create_structure():
    base_path = Path(os.getcwd())
    print(f"Initializing project structure at: {base_path}")

    # 1. Create Directories
    for folder, subfolders in structure.items():
        folder_path = base_path / folder
        folder_path.mkdir(exist_ok=True)
        print(f"Created: {folder}/")
        
        # Create .gitkeep to ensure git tracks empty folders
        (folder_path / ".gitkeep").touch()
        
        for sub in subfolders:
            sub_path = folder_path / sub
            sub_path.mkdir(exist_ok=True)
            (sub_path / ".gitkeep").touch()
            print(f"   └── Created: {folder}/{sub}/")

    # 2. Create Files
    for filename, content in files.items():
        file_path = base_path / filename
        if not file_path.exists():
            with open(file_path, "w") as f:
                f.write(content)
            print(f"Created: {filename}")
        else:
            print(f"Skipped: {filename} (already exists)")

    print("\nProject Structure Setup Complete!")
    print("Next Step: Run 'pip install -r requirements.txt' in your terminal.")

if __name__ == "__main__":
    create_structure()
```
